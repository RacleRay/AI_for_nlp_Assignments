````python
from sklearn.datasets import load_boston
import numpy as np
import random

data = load_boston()
x,  y = data['data'][:, 5],  data['target']
````

````python
def loss(y, y_hat):
    '''MSE'''
    return np.mean(np.power(y - y_hat, 2))
````

````python
def model(x, k, b):
    '''simple linear'''
    return x * k + b
````

#### random method

````python
def random_search(x, y, times, threshold=1e-3):
    best_k, best_b = None, None
    min_loss =  float('inf')
    for i in range(times):
        k = random.random() * 200 - 100
        b = random.random() * 200 - 100
        
        y_hat = model(x, k, b)
        c_loss = loss(y, y_hat)
        
        if c_loss < min_loss:
            min_loss = c_loss
            best_k, best_b = k, b
            print('When time is : {}, get best_k: {} best_b: {}, and the loss is: {}'.format(i, best_k, best_b, min_loss))            
    return best_k, best_b
````

#### Supervised Direction

````python
def direction_search(x, y, times, threshold=1e-3):
    # k, b
    direction = [
        (+1, -1),  
        (+1, +1), 
        (-1, -1), 
        (-1, +1),
    ]
    next_direc = random.choice(direction)
    scalar = 0.1
    
    k = random.random() * 200 - 100
    b = random.random() * 200 - 100
    
    best_k, best_b = None, None
    min_loss =  float('inf')
    update_times = 0
    
    for i in range(times):
        k_dir, b_dir = next_direc
        k, b = k+scalar*k_dir, b+scalar*b_dir
        
        y_hat = model(x, k, b)
        c_loss = loss(y, y_hat)
        
        if c_loss < min_loss:
            min_loss = c_loss
            best_k, best_b = k, b
            
            next_direc = next_direc
            update_times += 1
            if update_times%20 == 0:
                print('When time is : {}, get best_k: {} best_b: {}, and the loss is: {}'.format(i, best_k, best_b, min_loss))
        else:
            next_direc = random.choice(direction)
            
    return best_k, best_b
````

#### Gradient Descent and abs loss

$$Loss = \frac{1}{n} sum(|{y_i - \hat{y_i}}|)$$

````python
def loss_abs(y, y_hat):
    return np.mean(np.abs(y - y_hat))

def mask(y, y_hat):
    return np.sign(y - y_hat)

def a_partial_k(x, y, y_hat):
    return np.mean(- x * mask(y, y_hat))

def a_partial_b(y, y_hat):
    return np.mean(-mask(y, y_hat))
````

````python
def partial_k(x, y, y_hat):
    return -2 * np.mean((y - y_hat) * x)

def partial_b(y, y_hat):
    return -2 * np.mean(y - y_hat)

def gradient_descend(x, y, times, learning_rate, loss_fun, partial_k, partial_b, threshold=1e-3):
    k = random.random() * 200 - 100
    b = random.random() * 200 - 100
    
    best_k, best_b = None, None
    min_loss =  float('inf')
    
    for i in range(times):
        y_hat = model(x, k, b)
        c_loss = loss_fun(y, y_hat)
        
        if c_loss < min_loss:
            min_loss = c_loss
            best_k, best_b = k, b
        if i % 50 == 0: 
            print('When time is : {}, get best_k: {} best_b: {}, and the loss is: {}'.format(i, best_k, best_b, min_loss))
         
        k_gradient = partial_k(x, y, y_hat)
        b_gradient = partial_b(y, y_hat)
        
        k = k - learning_rate * k_gradient
        b = b - learning_rate * b_gradient
        
    return best_k, best_b
````

````python
from collections import defaultdict


def extract_sites(raw):
    return re.findall('>([\u4e00-\u9fa5|T2|T3]*?)<', raw)


def crawler(start_url):
    url = start_url
    header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.80 Safari/537.36'}
    content = requests.get(url, headers=header).content.decode('utf-8')
    
    pattern = re.compile('<li.*?class.*?<a.*?href=.*?class.*?>([\u4e00-\u9fa5 | \w | \S]*?)<(.*?)</li>', re.S)
    results = re.findall(pattern, content)
    results  = [i for i in results if '北京地铁' in i[0] if '未开通' not in i[1]]
    
    data = defaultdict()
    for res in results:
        data[res[0]] = extract_sites(res[1])
        if '环' in res[0]:
            data[res[0]].append(data[res[0]][0])
            
        
    return data
````

##### 3. Build the search agent

> Build the search agent based on the graph we build.

for example, when you run: 

```python
>>> search('奥体中心', '天安门') 
```
you need get the result: 

奥体中心-> A -> B -> C -> ... -> 天安门



````python
def search(start, destination):
    """
    breath first search
    graph: 相当于邻接表，存储邻接关系
    start：起点
    """
    pathes = [[start]]  # 所有路径
    visited = dict()  # 已遍历节点
    res = list()  # 查找结果
    global graph  # 声明图
    
    while pathes:
        path = pathes.pop(0)
        frontier = path[-1]  # 待遍历
        
        # 加入visited次数，防止每个节点只被遍历一次
        # 理论上说，路线是有很多的，相当于加了松弛变量
        if frontier in visited and visited[frontier] > 30: continue
        
        successors = graph[frontier]
        for site in successors:  # 遍历未知节点
            if site in path: continue
            new_path = path + [site]
            pathes.append(new_path)
            
            if site == destination:  # 记录结果
                res.append(new_path)
        
        if frontier not in visited:  # 记录已遍历节点
            visited[frontier] = 1  
        else:
            visited[frontier] += 1 
    
#     print(pathes)
    res = sorted(res, key=len)
    return res
````
